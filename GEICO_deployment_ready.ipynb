```python
#Creating copy of the data for manipulation
data_process= data.copy()
print(data_process.shape)
data_process.head()
```

```python
#Decided to drop the null values in the data since we only have 0.6% of missing data 
data_process.dropna(inplace=True)
print ("Data Shape:", data_process.shape, "\n")
```

```python
#Checking if data is null?
data_process.info(verbose=True)
```

```python
data_process = data_process.astype({"Tenure":'int', "Age":'int',"CHANNEL1_6M":'int', "CHANNEL2_6M":'int',"CHANNEL3_6M":'int', "CHANNEL4_6M":'int',"CHANNEL5_6M":'int', "METHOD1_6M":'int',"RECENT_PAYMENT":'int',"PAYMENTS_6M":'int'}) 
```

```python
data_process.describe()
```

```python
#Dividing the tenure into bucket__________________
bins_T = np.arange(data_process['Tenure'].min(),data_process['Tenure'].max() + 5, 5)
labels_T = ['-'.join(map(str,(x,y))) for x, y in zip(bins_T[:-1], bins_T[1:])]

data_process['Tenure_c'] = pd.cut(data_process['Tenure'], bins = bins_T, labels=labels_T)
```

```python
data_process['Tenure_c'].value_counts()
```

```python
#Dividing age into bucket__________________
bins_A = np.arange(data_process['Age'].min() - 5,data_process['Age'].max() + 5, 5)
labels_A = ['-'.join(map(str,(x,y))) for x, y in zip(bins_A[:-1], bins_A[1:])]

data_process['Age_c'] = pd.cut(data_process['Age'], bins = bins_A, labels=labels_A)
```

```python
data_process['Age_c'].value_counts()
```

```python
#Bins for logins__________
bins_L = np.arange(data_process['LOGINS'].min(),data_process['LOGINS'].max() + 5, 5)
labels_L = ['-'.join(map(str,(x,y))) for x, y in zip(bins_L[:-1], bins_L[1:])]

data_process['LOGINS_c'] = pd.cut(data_process['LOGINS'], bins = bins_L, labels=labels_L)
```

```python
data_process['LOGINS_c'].value_counts()
```

```python
#Dividing Rated State of Policy in 4 division based on the averaged Call_Flag.  
#RTD_ST_CD variable has 51 unique values, which seems to indicate states of the US.
#The reason is that getting 51 categorical dummies are creating way too many variables, so I'm going to be getting three categorical dummies instead.
#Then the columns DATE_FOR and RTD_ST_CD are dropped.
data_process=data_process.join(pd.cut(data_process.groupby(
            "RTD_ST_CD")["Call_Flag"].mean().sort_values(), 4, labels=[
            "0","1","2","3"]), on="RTD_ST_CD", rsuffix="_ST")
print(data_process.join(pd.cut(data_process.groupby(
            "RTD_ST_CD")["Call_Flag"].mean().sort_values(), 4, labels=[
            "0","1","2","3"]), on="RTD_ST_CD", rsuffix="_ST"))
```

```python
def uniplot(data,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(data[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = data, x= col, order=data[col].value_counts().index,hue = hue,palette='bright') 
        
    plt.show()
```

```python
## Pipelines Creation
## 1. Data Preprocessing by using Standard Scaler
## 2. Reduce Dimension using PCA
## 3. Apply  Classifier
```

```python
pipeline_lr=Pipeline([('scalar1',StandardScaler()),
                    ('pca1',PCA(n_components=2)),
                    ('lr_classifier',LogisticRegression(random_state=0))])
```

```python
pipeline_dt=Pipeline([('scalar2',StandardScaler()),
                     ('pca2',PCA(n_components=2)),
                     ('dt_classifier',DecisionTreeClassifier())])
```

```python
pipeline_randomforest=Pipeline([('scalar3',StandardScaler()),
                     ('pca3',PCA(n_components=2)),
                     ('rf_classifier',RandomForestClassifier())])
```

```python
# the list of pipelines
pipelines = [pipeline_lr, pipeline_dt, pipeline_randomforest]
```

```python
best_accuracy=0.0
best_classifier=0
best_pipeline=""
```

```python
# Dictionary of pipelines and classifier types for ease of reference
pipe_dict = {0: 'Logistic Regression', 1: 'Decision Tree', 2: 'RandomForest'}

# Fit the pipelines
for pipe in pipelines:
	pipe.fit(X_train, y_train)
```

```python
for i,model in enumerate(pipelines):
    print("{} Test Accuracy: {}".format(pipe_dict[i],model.score(X_test,y_test)))
```

```python
for i,model in enumerate(pipelines):
    if model.score(X_test,y_test)>best_accuracy:
        best_accuracy=model.score(X_test,y_test)
        best_pipeline=model
        best_classifier=i
print('Classifier with best accuracy:{}'.format(pipe_dict[best_classifier]))
```
